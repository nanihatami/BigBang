#### reading dataset file

```python
import pandas as pd
dataset = pd.read_csv("titanic.csv")
dataset
```

```python
dataset.isnull().sum()
dataset
```

# preprocessing

#### filling empty values

```python
dataset_age = dataset["Age"].fillna(dataset["Age"].mean())
del dataset["Age"]
dataset_embarked = dataset["Embarked"]
del dataset["Embarked"]
dataset = pd.concat([dataset, dataset_age, dataset_embarked], axis=1)
dataset
```

#### drop missing values

```python
del dataset["Cabin"]
```

```python
del dataset["PassengerId"]
```

```python
dataset.dropna(inplace=True, ignore_index=True)
dataset
```

#### one hot encoding

```python
for i in range(889):
    if dataset.loc[i, "Sex"] == "male":
        dataset.loc[i, "Sex"] = 1
    elif dataset.loc[i, "Sex"] == "female":
        dataset.loc[i, "Sex"] = 0
dataset
```

```python
one_hot_ticket = pd.get_dummies(dataset["Ticket"])
del dataset["Ticket"]
del dataset["Name"]
dataset_embarked = dataset["Embarked"]
del dataset["Embarked"]
dataset = pd.concat([dataset, one_hot_ticket, dataset_embarked], axis=1)
dataset
```

### scaling

```python
from sklearn.preprocessing import MinMaxScaler #,StandardScaler 
```

```python
minMax = MinMaxScaler()
# standard = StandardScaler()
dataset[["Survived", "Pclass"]] = minMax.fit_transform(dataset[["Survived", "Pclass"]])
dataset[["Sex", "SibSp", "Parch", "Fare", "Age"]] = minMax.fit_transform(dataset[["Sex", "SibSp", "Parch", "Fare", "Age"]])
dataset
```

# modeling

```python
from sklearn.model_selection import train_test_split
```

```python
features = dataset.iloc[:, :-1]
labels = dataset.iloc[:, -1]
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix#, recall_score, precision_score
```

### d-tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)
y_predict = tree.predict(x_test)
```

```python
print(accuracy_score(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
#print(recall_score(y_test, y_predict))
#print(precision_score(y_test, y_predict))
```

### svm

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm.fit(x_train, y_train)
y_predict = svm.predict(x_test)
```

```python
print(accuracy_score(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
```

### neural network

```python
from sklearn.neural_network import MLPClassifier
```

```python
model = MLPClassifier(hidden_layer_sizes=(30,31,31))
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
```

```python
print(accuracy_score(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
```
